```python
import pandas as pd

data=[]
# 텍스트 파일 줄바꿈 기준으로 읽기
with open('쿠팡정리2023.txt', 'r', encoding='utf-8') as file:
    lines = file.read().splitlines()  # 줄바꿈 기준으로 나눔
    for line in lines:
        data.append(line.split('\t'))  # 각 줄을 탭 기준으로 나눔

# 데이터를 DataFrame으로 변환 (예: 각 줄을 한 열로)
df = pd.DataFrame(data, columns=['주문일','주문ID','사업자코드','사업자명','상품ID','#','상품명','카테고리','가격'])
df.info() # 데이터프레임 전체 구조
```

```python
columns_to_drop = df.columns[[1,2,3,4,5,7]]
df = df.drop(columns=columns_to_drop)
print(df.head())
```

```python
import numpy as np

# 빈 문자열을 NaN으로 변환
df.replace("", np.nan, inplace=True)
```

```python
# 비어있는 행
df.isnull().sum()
```

```python
# NaN이 포함된 행 삭제
df.dropna(how='all', inplace=True)

print(df.head())  # 결과 확인
print(df.isnull().sum())
```

```python
# 엑셀 파일로 저장 
df.to_excel('coupang2023.xlsx', index=False)
```
